```python
import os

# Directory containing the annotation (.txt) files.
# Change this to your actual labels directory.
labels_dir = "/content/val/annotations"

# Mapping from file name prefix to the desired class label.
prefix_to_class = {
    "AF": "0",
    "AM": "1",
    "CF": "2",
    "CM": "3",
}

# Process each annotation file in the labels directory.
for filename in os.listdir(labels_dir):
    if not filename.endswith(".txt"):
        continue  # Skip non-text files

    file_path = os.path.join(labels_dir, filename)
    # Get the base file name (without extension) to check its prefix.
    base_name = os.path.splitext(filename)[0]

    # Determine the new class label based on the prefix.
    new_class = None
    for prefix, label in prefix_to_class.items():
        if base_name.startswith(prefix):
            new_class = label
            break

    if new_class is None:
        print(f"No matching prefix found for file: {filename}. Skipping.")
        continue

    new_lines = []
    with open(file_path, "r") as file:
        for line in file:
            parts = line.strip().split()
            # Check for valid annotation format (YOLO format: class x_center y_center width height)
            if len(parts) != 5:
                print(f"Skipping invalid line in {filename}: {line.strip()}")
                continue
            # Overwrite the class label with the new class from the file name.
            parts[0] = new_class
            new_lines.append(" ".join(parts))

    # Write the updated annotations back to the same file.
    with open(file_path, "w") as file:
        file.write("\n".join(new_lines) + "\n")

    print(f"Updated {filename} with class {new_class}")
```
